```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)].copy()

rows_removed = len(df) - len(df_cleaned)
print(f"Original rows: {len(df)}")
print(f"Rows after outlier removal: {len(df_cleaned)}")
print(f"Number of outliers removed: {rows_removed}")
```

## Part 1: Analyzing Best-Seller Trends Across Product Categories

1. Crosstab Analysis:

    - Create a crosstab between the product category and the isBestSeller status.

    - Are there categories where being a best-seller is more prevalent?

```python
proportion_crosstab = df_cleaned.groupby('category')['isBestSeller'].value_counts(normalize=True).mul(100).unstack(fill_value=0).round(2)

best_seller_proportion = proportion_crosstab[True].sort_values(ascending=False)

print("Top 10 Categories by Proportion of Best-Sellers (% of listings in that category are best-sellers):\n")
print(best_seller_proportion.head(10))
```

2. Statistical Tests:

    - Conduct a Chi-square test to determine if the best-seller distribution is independent of the product category.
    - Compute Cramér's V to understand the strength of association between best-seller status and category.

```python
top_10_categories = [
    'Smart Home Security & Lighting', 'Wind Instruments', 'Grocery',
    'Health & Personal Care', 'Material Handling Products',
    'Professional Medical Supplies', 'Billiard, Snooker & Pool',
    'Mobile Phone Accessories', 'Home Brewing & Wine Making',
    'Motorbike Exhaust & Exhaust Systems'
]

df_filtered = df_cleaned[df_cleaned['category'].isin(top_10_categories)].copy()

crosstab_result = pd.crosstab(df_filtered['category'], df_filtered['isBestSeller'])

from scipy.stats import chi2_contingency
from scipy.stats.contingency import association

chi2, p, dof, expected = chi2_contingency(crosstab_result)

cramers_v = association(crosstab_result, method="cramer")

print(f"Chi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.5f}")
print(f"Degrees of Freedom (dof): {dof}")
print(f"Cramér's V: {cramers_v:.3f}")
```

3. Visualizations:

    - Visualize the relationship between product categories and the best-seller status using a stacked bar chart.

```python
crosstab_proportions = pd.crosstab(
    df_filtered['category'], 
    df_filtered['isBestSeller'], 
    normalize='index'
)

plt.figure(figsize=(12, 7))
crosstab_proportions.plot(
    kind='bar', 
    stacked=True, 
    figsize=(12, 7),
    color=['skyblue', 'orange']
)

plt.title('Proportion of Best Sellers by Top 10 Product Category')
plt.xlabel('Product Category')
plt.ylabel('Proportion (Normalized to 1.0)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Is Best Seller', labels=['False', 'True'])
plt.tight_layout()
plt.savefig('stacked_bar_chart_bestseller.png')
plt.show()
```

## Part 2: Exploring Product Prices and Ratings Across Categories and Brands

1. Violin Plots:

    - Use a violin plot to visualize the distribution of price across different product categories. Filter out the top 20 categories based on count for better visualization.
    - Which product category tends to have the highest median price? Don't filter here by top categories.


```python
category_counts = df_cleaned['category'].value_counts()

top_20_categories = category_counts.head(20).index.tolist()

df_top_20 = df_cleaned[df_cleaned['category'].isin(top_20_categories)].copy()
```

```python
plt.figure(figsize=(14, 8))
sns.violinplot(
    data=df_top_20,
    x='category',
    y='price',
    palette='Set3'
)

plt.title('Distribution of Price across Top 20 Product Categories (Outliers Removed)')
plt.xlabel('Product Category')
plt.ylabel('Price (£)')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
median_prices = df_cleaned.groupby('category')['price'].median()

highest_median_price_category = median_prices.idxmax()
max_median_price = median_prices.max()

print(f"Category with the Highest Median Price (across all categories): {highest_median_price_category}")
print(f"Highest Median Price: {max_median_price:.2f}")
```

2. Bar Charts:

    - Create a bar chart comparing the average price of products for the top 10 product categories (based on count).
    - Which product category commands the highest average price? Don't filter here by top categories.

```python
category_counts_cleaned = df_cleaned['category'].value_counts()
top_10_categories_count = category_counts_cleaned.head(10).index.tolist()

df_top_10 = df_cleaned[df_cleaned['category'].isin(top_10_categories_count)].copy()

average_prices_top_10 = df_top_10.groupby('category')['price'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 7))
sns.barplot(
    x=average_prices_top_10.index,
    y=average_prices_top_10.values,
    palette='viridis'
)

plt.title('Average Price for Top 10 Most Listed Categories (Outliers Removed)')
plt.xlabel('Product Category')
plt.ylabel('Average Price (£)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

average_prices_all = df_cleaned.groupby('category')['price'].mean()
highest_avg_price_category = average_prices_all.idxmax()
max_avg_price = average_prices_all.max()

print(f"Category with the Highest Average Price (across all categories): {highest_avg_price_category}")
print(f"Highest Average Price: {max_avg_price:.2f}")
```

3. Box Plots:

    - Visualize the distribution of product ratings based on their category using side-by-side box plots. Filter out the top 10 categories based on count for better visualization.
    - Which category tends to receive the highest median rating from customers? Don't filter here by top categories.

```python
category_counts_cleaned = df_cleaned['category'].value_counts()
top_10_categories_count = category_counts_cleaned.head(10).index.tolist()

df_top_10_ratings = df_cleaned[df_cleaned['category'].isin(top_10_categories_count)].copy()

median_ratings_all = df_cleaned.groupby('category')['stars'].median()
highest_median_rating_category = median_ratings_all.idxmax()
max_median_rating = median_ratings_all.max()

plt.figure(figsize=(12, 7))
sns.boxplot(
    data=df_top_10_ratings,
    x='category',
    y='stars',
    palette='Pastel1'
)

plt.title('Distribution of Product Ratings across Top 10 Categories')
plt.xlabel('Product Category')
plt.ylabel('Rating (Stars)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(f"Category with the Highest Median Rating (across all categories): {highest_median_rating_category}")
print(f"Highest Median Rating: {max_median_rating:.2f}")
```

## 3:Investigating the Interplay Between Product Prices and Ratings

1. Correlation Coefficients:

    - Calculate the correlation coefficient between price and stars.
    - Is there a significant correlation between product price and its rating?

```python
pearson_corr = df_cleaned['price'].corr(df_cleaned['stars'], method='pearson')

spearman_corr = df_cleaned['price'].corr(df_cleaned['stars'], method='spearman')

print(f"Pearson Correlation (Linear): {pearson_corr:.3f}")
print(f"Spearman Correlation (Monotonic): {spearman_corr:.3f}")
```

Both the Pearson Correlation (-0.078) and the Spearman Correlation (-0.067$) are very close to zero.

- Strength: The magnitude of both coefficients is low (below 0.1), indicating a very weak or negligible relationship between product price and customer rating.

- Significance: The low value suggests there is no meaningful correlation between the price you pay for a product and the star rating it receives.

In this market, paying more does not linearly guarantee a better-rated product!

2. Visualizations:

    - Use a scatter plot to visualize the relationship between product rating and price. What patterns can you observe?
    - Use a correlation heatmap to visualize correlations between all numerical variables.
    - Examine if product prices typically follow a normal distribution using a QQ plot.

```python
df_cleaned['log_price'] = np.log1p(df_cleaned['price'])

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='stars', y='log_price', alpha=0.3)
plt.title('Log-Transformed Price vs. Product Ratings (Outliers Removed)')
plt.xlabel('Rating (Stars)')
plt.ylabel('Log(Price + 1)')
plt.show()
```
